# Most completed solves in each event

```python
df1 = df[['personId','personName','personCountryId','eventId','value1','value2','value3','value4','value5']]

result = pd.DataFrame()

solves = pd.concat([
    df1[df1['value1']>0][['eventId', 'personId', 'personName', 'value1']].rename(columns = {'value1':'value'}), 
    df1[df1['value2']>0][['eventId', 'personId', 'personName', 'value2']].rename(columns = {'value2':'value'}), 
    df1[df1['value3']>0][['eventId', 'personId', 'personName', 'value3']].rename(columns = {'value3':'value'}), 
    df1[df1['value4']>0][['eventId', 'personId', 'personName', 'value4']].rename(columns = {'value4':'value'}), 
    df1[df1['value5']>0][['eventId', 'personId', 'personName', 'value5']].rename(columns = {'value5':'value'})
    ], axis = 0)
    
solves['count'] = solves.groupby(['personId','eventId'])['value'].transform('count') 
solves = solves.drop_duplicates(subset = ['personId','eventId','count'], keep = 'first')
solves = solves.drop('value', 1)

for e in eventi:
    buff = solves[solves['eventId'] == e].sort_values(by = 'count', ascending = False).reset_index(drop = True)
    buff = buff.head(buff[buff['count'] == buff['count'].max()].shape[0])
    
    result = pd.concat([result, buff], axis = 0)
```

```python
result = result.sort_values(by = 'eventId')
result
```

```python
# display time

def timeconvert(x):
    if x < 60:
        return str(x)
    
    if x < 3600:
        m = x // 60
        return str(m)+'m '+str(x-(m*60))+'s'
    
    if x < 3600*24:
        h = x // 3600
        m = (x - h*3600) // 60
        s = (x - (h*3600+m*60))
        return str(h)+'h '+str(m)+'m '+ str(s)+'s'
    
    d = x // (24*3600)
    h = (x - (d*24*3600)) // 3600
    m = (x - (d*24*3600+h*3600)) // 60
    s = (x - (d*24*3600+h*3600+m*60))
    
    return str(d)+'d '+str(h)+'h '+ str(m)+'m '+str(s)+'s'
        
    
```
